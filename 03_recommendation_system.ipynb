```python
df['TotalCharges'] = pd.to_numeric(df['TotalCharges'], errors='coerce')  # convert to float
df.dropna(subset=['TotalCharges'], inplace=True)  # remove rows where conversion failed
```

```python
print(df[['tenure', 'MonthlyCharges', 'TotalCharges']].dtypes)
```

```python
from sklearn.preprocessing import StandardScaler
from sklearn.metrics.pairwise import cosine_similarity

features = ['tenure', 'MonthlyCharges', 'TotalCharges']

scaled = StandardScaler().fit_transform(df[features])
similarity = cosine_similarity(scaled)
```

```python
customer_id = 0  # pick a random customer
similar_customers = similarity[customer_id].argsort()[::-1][1:6]  # Top 5 similar
df.iloc[similar_customers]
```
